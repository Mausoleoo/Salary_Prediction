```python
import pandas as pd
```

```python
df = pd.read_csv('data scientist.csv')
```

```python
df.shape
```

```python
df.head()
```

### Remove empty Salary values

```python
df = df[df['salary estimate'] != '-1']
```

```python
df.shape
```

```python
df
```

### Salary parsing

#### Remove text

```python
salary = df['salary estimate'].apply(lambda x:x.split('(')[0])
```

```python
df['salary estimate'] = salary
```

```python
df
```

#### Divide by hour and year

```python
df['hourly'] = df['salary estimate'].apply(lambda x:1 if 'hr' in x.lower() else 0)
```

```python
df.head()
```

#### Remove special characters

```python
df['salary estimate'] = df['salary estimate'].apply(lambda x: x.lower().replace("$",""))
```

```python
df.head()
```

### Remove /hr and /yr

```python
df['salary estimate'] = df['salary estimate'].apply(lambda x:x.split('/')[0].replace(" ",""))
```

```python
df.head()
```

### Company name text only

#### Only Text

```python
df['rating'] = df['company'].apply(lambda x:x[-3:] if x[-2] == "." else -1)
```

#### Rating Column

```python
df['company'] = df['company'].apply(lambda x:x[:-3] if x[-2] == "." else x)
```

#### Remove break line

```python
df['company'] = df['company'].apply(lambda x:x.replace("\n",""))
```

```python
df.head()
```

### Remove Cities, Only States

```python
df['state'] = df['location'].apply(lambda x:x.split(',')[1] if len(x.split(',')) > 1 else x ) 
df['city'] = df['location'].apply(lambda x:x.split(',')[0])
```

```python
df.head()
```

```python
df.state.value_counts()
```

### Age Company

```python
df['age'] = df['company_founded'].apply(lambda x:-1 if x < 0 else 2023 - x)
```

```python
df.head()
```

### Parsing of job description

#### Most popular data science tools

```python
#Python
df['python_needed'] = df['job description'].apply(lambda x:1 if 'python' in x.lower() else 0)
#SAS
df['sas_needed'] = df['job description'].apply(lambda x:1 if 'sas' in x.lower() else 0)
#tableau
df['tableau_needed'] = df['job description'].apply(lambda x:1 if 'tableau' in x.lower() else 0)
#Matlab
df['matlab_needed'] = df['job description'].apply(lambda x:1 if 'matlab' in x.lower() else 0)
#Spark
df['spark_needed'] = df['job description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
#Tensorflow
df['tensorflow_needed'] = df['job description'].apply(lambda x:1 if 'tensorflow' in x.lower() else 0)
#Pytorch
df['pytorch_needed'] = df['job description'].apply(lambda x:1 if 'pytorch' in x.lower() else 0)
#Hadoop
df['hadoop_needed'] = df['job description'].apply(lambda x:1 if 'hadoop' in x.lower() else 0)
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis  = 1)
```
