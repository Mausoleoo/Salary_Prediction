### Salary parsing

#### Remove text

```python
salary = df['salary estimate'].apply(lambda x:x.split('(')[0])
```

```python
df['salary estimate'] = salary
```

```python
df
```

#### Divide by hour and year

```python
df['hourly'] = df['salary estimate'].apply(lambda x:1 if 'hr' in x.lower() else 0)
```

```python
df.head()
```

#### Remove special characters

```python
df['salary estimate'] = df['salary estimate'].apply(lambda x: x.lower().replace("$",""))
```

```python
df.head()
```

### Remove /hr and /yr

```python
df['salary estimate'] = df['salary estimate'].apply(lambda x:x.split('/')[0].replace(" ",""))
```

```python
df.head()
```

### Company name text only

#### Only Text

```python
df['rating'] = df['company'].apply(lambda x:x[-3:] if x[-2] == "." else -1)
```

#### Rating Column

```python
df['company'] = df['company'].apply(lambda x:x[:-3] if x[-2] == "." else x)
```

#### Remove break line

```python
df['company'] = df['company'].apply(lambda x:x.replace("\n",""))
```

### Remove Cities, Only States

### Seniority

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return ' data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning engineer'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'

def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'junior':
        return 'junior'
    else:
        return 'na'
```

```python
df['job_simp'] = df['job title'].apply(title_simplifier)
```

```python
df['seniority'] = df['job title'].apply(seniority)
```
